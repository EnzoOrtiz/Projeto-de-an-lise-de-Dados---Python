Gráfico 1: Total de Vagas Autorizadas por Região


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("PDA_Dados_Cursos_Graduacao_Brasil.csv")

vagas_por_regiao = df.groupby('REGIAO')['QT_VAGAS_AUTORIZADAS'].sum()

plt.figure(figsize=(12, 6)) 

plt.bar(vagas_por_regiao.index, vagas_por_regiao.values, color='skyblue')

plt.title('Total de Vagas Autorizadas por Região', fontsize=16)
plt.xlabel('Região', fontsize=12)
plt.ylabel('Total de Vagas', fontsize=12)
plt.grid(True)

plt.savefig("Gráfico_vagasporregiao.png")
plt.show()
```

Gráfico 2: Modalidades de Ensino por Região (EAD vs Presencial)


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("PDA_Dados_Cursos_Graduacao_Brasil.csv")

modalidade_por_regiao = df.groupby(['REGIAO', 'MODALIDADE']).size().unstack()

plt.figure(figsize=(12, 6))  

modalidade_por_regiao.plot(kind='bar', stacked=True, color=['#4CAF50', '#FF9800'], ax=plt.gca())

plt.title('Modalidades de Ensino por Região (EAD vs Presencial)', fontsize=16)
plt.xlabel('Região', fontsize=12)
plt.ylabel('Número de Cursos', fontsize=12)
plt.grid(True, axis='y')
plt.xticks(rotation=0)

plt.savefig("Gráfico_modalidade.png")
plt.show()
```

Gráfico 3: Vagas Autorizadas por Curso (Primeiros 10 Cursos)


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("PDA_Dados_Cursos_Graduacao_Brasil.csv")

cursos = df['NOME_CURSO'].head(10)
vagas_autorizadas = df['QT_VAGAS_AUTORIZADAS'].head(10)

plt.figure(figsize=(12, 6))  

plt.plot(cursos, vagas_autorizadas, marker='o', linestyle='-', color='blue')

plt.title('Vagas Autorizadas por Curso', fontsize=16)
plt.xlabel('Cursos', fontsize=12)
plt.ylabel('Quantidade de Vagas Autorizadas', fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=10)  
plt.grid(True)

plt.savefig("Gráfico_vagascurso.png", bbox_inches='tight')
plt.show()
```

Gráfico 4: Carga Horária Total por Instituição (Primeiras 10 Instituições)


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("PDA_Dados_Cursos_Graduacao_Brasil.csv")

carga_horaria_por_instituicao = df.groupby('NOME_IES')['CARGA_HORARIA'].sum().head(10)

plt.barh(carga_horaria_por_instituicao.index, carga_horaria_por_instituicao.values, color='lightgreen')
plt.title('Carga Horária Total por Instituição')
plt.xlabel('Carga Horária Total')
plt.ylabel('Instituição')
plt.grid(True)

plt.savefig("Gráfico_cargahoraria.png", bbox_inches='tight')
plt.show()
```
